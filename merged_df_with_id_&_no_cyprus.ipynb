```python
#import pandas library and the dataset
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

merged_df = pd.read_csv('/content/drive/My Drive/happiness_project/whr/merged_WHR.csv')
```

```python
#check the length of the total dataframe
len(merged_df)
```

```python
# check the number of row containing Cyprus as country
cyprus_df = merged_df[merged_df["country"] == "cyprus"]
len(cyprus_df)
```

```python
#just to have a look at the lines containing Cyprus
cyprus_df
```

```python
# remove Cyprus from the dataframe
merged_df = merged_df[merged_df["country"] != "cyprus"]
```

```python
len(merged_df)
```

```python
#creation of an unique id
merged_df['id'] = merged_df['country'].astype(str) + '_' + merged_df['year'].astype(str)
```

```python
merged_df_w_id = merged_df
```
