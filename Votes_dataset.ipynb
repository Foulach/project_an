```python
import pandas as pd

# Assuming 'merged_df' is already loaded

# Group by 'titre' then 'groupe politique', then count votes
votes_by_title_and_affiliation = merged_df.groupby(['titre', 'groupe politique'])[['pour', 'pours', 'contre', 'contres', 'abstention', 'abstentions']].sum()

# Reset index to move both 'titre' and 'groupe politique' back to columns
votes_by_title_and_affiliation = votes_by_title_and_affiliation.reset_index()

# Filter for Renaissance votes
renaissance_votes = votes_by_title_and_affiliation[votes_by_title_and_affiliation['groupe politique'] == 'Renaissance']

# Calculate agreement and disagreement scores for each group
agreement_scores = {}
for group in votes_by_title_and_affiliation['groupe politique'].unique():
    if group == 'Renaissance':
        continue  # Skip Renaissance itself

    group_votes = votes_by_title_and_affiliation[votes_by_title_and_affiliation['groupe politique'] == group]

    # Calculate agreement (same vote type) and disagreement (opposite vote type)
    agreement = 0
    disagreement = 0
    for _, renaissance_row in renaissance_votes.iterrows():
        for _, group_row in group_votes.iterrows():
            if renaissance_row['titre'] == group_row['titre']:
                # Access columns individually, sum the votes
                if (renaissance_row['pour'] + renaissance_row['pours'] > renaissance_row['contre'] + renaissance_row['contres'] and
                    group_row['pour'] + group_row['pours'] > group_row['contre'] + group_row['contres']) or \
                   (renaissance_row['contre'] + renaissance_row['contres'] > renaissance_row['pour'] + renaissance_row['pours'] and
                    group_row['contre'] + group_row['contres'] > group_row['pour'] + group_row['pours']):
                    agreement += 1
                else:
                    disagreement += 1

    agreement_scores[group] = {'agreement': agreement, 'disagreement': disagreement}

# Print the results
for group, scores in agreement_scores.items():
    print(f"Groupe: {group}, Agreement: {scores['agreement']}, Disagreement: {scores['disagreement']}")
```
